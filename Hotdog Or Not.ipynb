```python
def get_image_vector(img_path, dims):
    img_arr = np.array(mpimg.imread(img_path, img_path.split(".")[-1]))
    img = Image.fromarray(img_arr)
    img = img.resize(dims, Image.ANTIALIAS)
    img_vec = np.array(img)
    img_vec = img_vec / 255
    img_vec.reshape(255*255*3,)
    print(img_vec.shape)
```
